```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('size_MLP.csv', header = None)
data.columns = ['acc_test', 'acc_train', 'informative_samples', 'beta', 'size', 'hidden_size']
```

```python
data
```

```python
size = 50000 
informative_samples = 5000
data_filter = data[data['size']==size]
#data_filter = data_filter[data_filter['hidden_size']==100]
data_filter = data_filter[data_filter['informative_samples']==informative_samples]
data_filter = data_filter[data_filter['beta']==-1]
```

```python
t = np.arange(0., 0.11, 0.01)
plt.scatter(data_filter['hidden_size'], data_filter['acc_train'],  label='Accuracy training', marker='x')
plt.scatter(data_filter['hidden_size'], data_filter['acc_test'], label='Accuracy testing', marker='>')
#plt.plot(t,[86 for i in range(11)], 'r--', label = 'without samples')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.ylim(40, 100)
#plt.xlim(0, 0.1)
plt.ylabel('Accuracy (%)')
plt.xlabel('Hidden size')
plt.savefig('images/'+str(size)+'_'+str(informative_samples)+"_Loss_Training.png")
```

```python
t = np.arange(0., 0.11, 0.01)
plt.scatter(data_filter['beta'], data_filter['acc_train'],  label='Accuracy training')
plt.scatter(data_filter['beta'], data_filter['acc_test'], label='Accuracy testing')
plt.plot(t,[86 for i in range(11)], 'r--', label = 'without samples')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.ylim(60, 100)
plt.xlim(0, 0.1)
plt.ylabel('Accuracy (%)')
plt.xlabel('beta')
plt.savefig('images/'+str(size)+'_'+str(informative_samples)+"_Loss_Training.png")
```
